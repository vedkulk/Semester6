```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.histplot(new_df["person_age"], bins=30, kde=True)
plt.title("Age Distribution of Loan Applicants")
plt.show()


plt.figure(figsize=(8, 5))
sns.scatterplot(x="loan_amnt", y="loan_int_rate", hue="loan_status", data=new_df)
plt.title("Loan Amount vs Interest Rate")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="loan_status", y="credit_score", data=new_df)
plt.title("Credit Score Distribution by Loan Status")
plt.xlabel("Loan Status")
plt.ylabel("Credit Score")
plt.show()
```
